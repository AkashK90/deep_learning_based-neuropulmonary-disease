```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.applications import VGG16  # Using VGG16 instead of EfficientNet (me)
import cv2
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

# Global variables (students often use these instead of class attributes)
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 50  # Reduced epochs

class MedicalClassifier:
    def __init__(self):
        # Simple initialization without complex config
        self.brain_classes = ['Glioma', 'Meningioma', 'Normal', 'Pituitary']
        self.chest_classes = ['Normal', 'Pneumonia']
        self.brain_model = None
        self.chest_model = None

    def load_image(self, image_path):
        # Basic image loading function
        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img.astype('float32') / 255.0
        return img

    def create_brain_model(self):
        # Simple CNN model (student-level architecture)
        model = Sequential()
        model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)))
        model.add(MaxPool2D(2, 2))
        model.add(Conv2D(64, (3, 3), activation='relu'))
        model.add(MaxPool2D(2, 2))
        model.add(Conv2D(128, (3, 3), activation='relu'))
        model.add(MaxPool2D(2, 2))
        model.add(Conv2D(256, (3, 3), activation='relu'))
        model.add(MaxPool2D(2, 2))

        model.add(Flatten())
        model.add(Dense(512, activation='relu'))
        model.add(Dropout(0.5))
        model.add(Dense(256, activation='relu'))
        model.add(Dropout(0.3))
        model.add(Dense(4, activation='softmax'))  # 4 classes for brain

        model.compile(optimizer=Adam(learning_rate=0.001),
                     loss='categorical_crossentropy',
                     metrics=['accuracy'])
        return model
        model.summary()

    def create_chest_model(self):
        # Using pre-trained VGG16 (common choice for me)  in this model  https://www.geeksforgeeks.org/computer-vision/vgg-16-cnn-model/
        base_model = VGG16(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
        base_model.trainable = False  # Freeze base model

        model = Sequential([
            base_model,
            GlobalAveragePooling2D(),
            Dense(128, activation='relu'),
            Dropout(0.5),
            Dense(64, activation='relu'),
            Dropout(0.3),
            Dense(2, activation='softmax')  # 2 classes for chest
        ])

        model.compile(optimizer=Adam(learning_rate=0.001),
                     loss='categorical_crossentropy',
                     metrics=['accuracy'])
        return model
        model.summary()
    def train_brain_model(self, data_path):
        print("Training Brain MRI Model...")

        # Create data generators
        train_datagen = ImageDataGenerator(
            rescale=1.0/255,
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            validation_split=0.2
        )

        # Load training data
        train_generator = train_datagen.flow_from_directory(
            os.path.join(data_path, 'Training'),
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            subset='training'
        )

        val_generator = train_datagen.flow_from_directory(
            os.path.join(data_path, 'Training'),
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            class_mode='categorical',
            subset='validation'
        )

        # Create and train model
        self.brain_model = self.create_brain_model()

        # Simple callbacks
        callbacks = [
            ModelCheckpoint('brain_model.h5', save_best_only=True, monitor='val_accuracy'),
            EarlyStopping(monitor='val_accuracy', patience=8, restore_best_weights=True)
        ]

        # Train model
        history = self.brain_model.fit(
            train_generator,
            epochs=EPOCHS,
            validation_data=val_generator,
            callbacks=callbacks
        )

        # Simple plotting
        self.plot_history(history, 'Brain MRI')
        return history

    def train_chest_model(self, data_path):
        print("Training Chest X-Ray Model...")

        # Similar structure but for chest data
        train_datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=15,
            width_shift_range=0.1,
            height_shift_range=0.1,
            zoom_range=0.1,
            horizontal_flip=True
        )

        val_datagen = ImageDataGenerator(rescale=1./255)

        train_generator = train_datagen.flow_from_directory(
            os.path.join(data_path, 'train'),
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            class_mode='categorical'
        )

        val_generator = val_datagen.flow_from_directory(
            os.path.join(data_path, 'test'),
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            class_mode='categorical'
        )

        # Create and train model
        self.chest_model = self.create_chest_model()

        callbacks = [
            ModelCheckpoint('chest_model.h5', save_best_only=True, monitor='val_accuracy'),
            EarlyStopping(monitor='val_accuracy', patience=8, restore_best_weights=True)
        ]

        history = self.chest_model.fit(
            train_generator,
            epochs=EPOCHS,
            validation_data=val_generator,
            callbacks=callbacks
        )

        self.plot_history(history, 'Chest X-Ray')
        return history

    def plot_history(self, history, title):
        # Basic plotting function
        plt.figure(figsize=(12, 4))

        # Accuracy plot
        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'], label='Training Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title(f'{title} - Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()

        # Loss plot
        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title(f'{title} - Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()

        plt.tight_layout()
        plt.savefig(f'{title.replace(" ", "_").lower()}_training.png')
        plt.show()

    def load_models(self):
        # Load saved models
        try:
            if os.path.exists('brain_model.h5'):
                self.brain_model = load_model('brain_model.h5')
                print("Brain model loaded!")
            if os.path.exists('chest_model.h5'):
                self.chest_model = load_model('chest_model.h5')
                print("Chest model loaded!")
        except Exception as e:
            print(f"Error loading models: {e}")

    def predict_brain(self, image_path):
        if self.brain_model is None:
            print("Brain model not loaded!")
            return None

        img = self.load_image(image_path)
        img = np.expand_dims(img, axis=0)

        prediction = self.brain_model.predict(img)
        class_idx = np.argmax(prediction[0])
        confidence = prediction[0][class_idx]

        result = {
            'diagnosis': self.brain_classes[class_idx],
            'confidence': float(confidence),
            'probabilities': {}
        }

        # Add all probabilities
        for i, class_name in enumerate(self.brain_classes):
            result['probabilities'][class_name] = float(prediction[0][i])

        return result

    def predict_chest(self, image_path):
        if self.chest_model is None:
            print("Chest model not loaded!")
            return None

        img = self.load_image(image_path)
        img = np.expand_dims(img, axis=0)

        prediction = self.chest_model.predict(img)
        class_idx = np.argmax(prediction[0])
        confidence = prediction[0][class_idx]

        result = {
            'diagnosis': self.chest_classes[class_idx],
            'confidence': float(confidence),
            'probabilities': {}
        }

        for i, class_name in enumerate(self.chest_classes):
            result['probabilities'][class_name] = float(prediction[0][i])

        return result

# Simple helper functions
def setup_directories():
    # Create basic directory structure
    directories = ['models', 'results']
    for directory in directories:
        os.makedirs(directory, exist_ok=True)
    print("Directories created!")

def get_user_paths():
    # Get dataset paths from user
    print("Please provide dataset paths:")
    brain_path = input("Brain tumor dataset path: ").strip()
    chest_path = input("Chest X-ray dataset path: ").strip()

    # Basic validation
    if brain_path and not os.path.exists(brain_path):
        print(f"Brain dataset not found: {brain_path}")
        brain_path = None

    if chest_path and not os.path.exists(chest_path):
        print(f"Chest dataset not found: {chest_path}")
        chest_path = None

    return brain_path, chest_path

def main():
    print("Medical Scan Classification System")
    print("-" * 10)

    # Setup
    setup_directories()
    classifier = MedicalClassifier()

    while True:
        print("\nOptions:")
        print("1. Train Brain MRI Model")
        print("2. Train Chest X-Ray Model")
        print("3. Test Models")
        print("4. Exit")

        choice = input("Enter choice (1-4): ").strip()

        if choice == '1':
            brain_path = input("Enter brain dataset path: ").strip()
            if brain_path and os.path.exists(brain_path):
                try:
                    classifier.train_brain_model(brain_path)
                    print("Brain model training completed!")
                except Exception as e:
                    print(f"Training failed: {e}")
            else:
                print("Invalid path!")

        elif choice == '2':
            chest_path = input("Enter chest dataset path: ").strip()
            if chest_path and os.path.exists(chest_path):
                try:
                    classifier.train_chest_model(chest_path)
                    print("Chest model training completed!")
                except Exception as e:
                    print(f"Training failed: {e}")
            else:
                print("Invalid path!")

        elif choice == '3':
            classifier.load_models()

            test_image = input("Enter test image path: ").strip()
            if not os.path.exists(test_image):
                print("Image not found!")
                continue

            print("Select scan type:")
            print("1. Brain MRI")
            print("2. Chest X-Ray")
            scan_type = input("Enter choice (1-2): ").strip()

            if scan_type == '1':
                result = classifier.predict_brain(test_image)
                if result:
                    print(f"Diagnosis: {result['diagnosis']}")
                    print(f"Confidence: {result['confidence']:.3f}")
                    print("All probabilities:")
                    for class_name, prob in result['probabilities'].items():
                        print(f"  {class_name}: {prob:.3f}")

            elif scan_type == '2':
                result = classifier.predict_chest(test_image)
                if result:
                    print(f"Diagnosis: {result['diagnosis']}")
                    print(f"Confidence: {result['confidence']:.3f}")
                    print("All probabilities:")
                    for class_name, prob in result['probabilities'].items():
                        print(f"  {class_name}: {prob:.3f}")

            else:
                print("Invalid choice!")

        elif choice == '4':
            print("Thanks for using the system!")
            break

        else:
            print("Invalid choice!")

if __name__ == "__main__":
    # Check TensorFlow
    print(f"TensorFlow version: {tf.__version__}")
    if tf.config.list_physical_devices('GPU'):
        print("GPU available!")
    else:
        print("Using CPU")

    # Run main program
    main()
```
